```python
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
import os

print("Everything working")
```

```python
# Define the path to the image
folder_path = "C:\\Users\\Lenovo\\OneDrive\\Desktop\\heart_rate_project\\sample_data\\Subject1\\NIR-002"
```

```python
# Read All pgm Files in the folder
files = sorted(os.listdir(folder_path))

print("Total number of files: ", len(files))

print("First 5 files: ", files[:5])
```

```python
# Load the first image to check 
img_path = os.path.join(folder_path, files[0])
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

print("Image shape: ", img.shape)
print("Image dtype: ", img.dtype)
```

```python
# Display the image
plt.imshow(img, cmap= 'gray')
plt.title("Sample NIR Image")
plt.axis('off')

plt.show()

```

```python
# Resize the image to 224x224
img_resized = cv2.resize(img, (64, 64))

print("Resized image shape: " , img_resized.shape)
```

```python
# print the resized image
plt.imshow(img_resized, cmap = 'gray')
plt.title("Resized NIR Image")
plt.axis('off')
plt.show()
```

```python
# Normalise the Image
img_normalized = img_resized / 255.0

print(img_normalized.min(), img_normalized.max())
```
